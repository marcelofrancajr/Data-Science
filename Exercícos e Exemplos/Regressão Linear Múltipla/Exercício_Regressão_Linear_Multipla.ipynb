```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Importe o arquivo formato xlsx
lucro = pd.read_excel("Exercicio_Lucro.xlsx")
lucro.head()
```

```python
#Tire a correlação entre as variáveis e analise
lucro.corr()
```

```python
#Faça o comparativo entre as variáveis
sns.pairplot(lucro);
```

```python
#Crie e analise os resultados da OLS e faça os ajustes necessários
import statsmodels.formula.api as smf
#modelo MRLS
function = 'Lucro~Investimento+Investimento_Mkt+Gastos_Administrativos'
model = smf.ols(formula=function, data=lucro).fit() 
print(model.summary())
```

```python
function = 'Lucro~Investimento+Investimento_Mkt'
model = smf.ols(formula=function, data=lucro).fit() 
print(model.summary())
```
